```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
```

```python
df = pd.read_csv("Indian Peninsula.csv")
print(df)
```

```python
df
```

```python
df['year'] = pd.DatetimeIndex(df['Start Date']).year
df['month'] = pd.DatetimeIndex(df['Start Date']).month
```

```python
df
```

```python
value_month = df['month'].value_counts()
print(value_month)
```

```python
y=[75, 69, 69]
x=['March',
 'April',
 'May']
plot.barh(x,y,color=['red', 'skyblue', 'skyblue'])
plot.title('Pre Monsoon Monthly distribution of flood events during 1965-2019')
plot.xlabel("Number of Flood Events")
plot.ylabel("Month")
for index, value in enumerate(y):
    plot.text(value, index, str(value))
```
